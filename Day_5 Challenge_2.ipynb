```python
text= ''''''
```

```python
def is_same_type(letter1,letter2):
    return(letter1.lower()==letter2.lower())
def opposite_polarity(letter1,letter2):
    if letter1!=letter2:
        if letter1==letter2.lower() or letter2==letter1.lower():
            return(True)
    return(False)
```

```python
has_reaction = True

while has_reaction:
    has_reaction = False
    #iterate through center of string
    for current_index in range(0,len(text)-2):
       
        if is_same_type(text[current_index],text[current_index+1])\
            and opposite_polarity(text[current_index],text[current_index+1]):
            #remove current and next character
            text = text[:current_index]+text[current_index+2:]
            has_reaction = True
            break

print(len(text) )    
```

```python
#print(len(text))
unique_characters = ''.join(list(set(text)))
#print(unique_characters)
lower = [x.lower() for x in unique_characters]
texts = {}
for character in lower:
    texts.update({character:text.replace(character,'').replace(character.upper(),'')})
print(texts.keys())
max_count = len(text)
for character in texts.keys():
    current_text = texts[character]
    has_reaction = True

    while has_reaction:
        has_reaction = False
        #iterate through center of string
        for current_index in range(1,len(current_text)-2):
            if is_same_type(current_text[current_index],current_text[current_index+1])\
                and opposite_polarity(current_text[current_index],current_text[current_index+1]):
                #remove current and next character
                current_text = current_text[:current_index]+current_text[current_index+2:]
                has_reaction = True
                break
            elif is_same_type(current_text[current_index],current_text[current_index-1])\
                and opposite_polarity(current_text[current_index],current_text[current_index-1]):
                #remove current and next character
                current_text = current_text[:current_index]+current_text[current_index+2:]
                has_reaction = True
                break

    if len(current_text) < max_count:
        max_count = len(current_text)
        print(character,'\t',len(current_text))
        
```
