```python
text= ''''''
```

```python
def is_same_type(letter1,letter2):
    return(letter1.lower()==letter2.lower())
def opposite_polarity(letter1,letter2):
    if letter1!=letter2:
        if letter1==letter2.lower() or letter2==letter1.lower():
            return(True)
    return(False)
```

```python
has_reaction = True

while has_reaction:
    has_reaction = False
    #iterate through center of string
    for current_index in range(0,len(text)-2):
       
        if is_same_type(text[current_index],text[current_index+1])\
            and opposite_polarity(text[current_index],text[current_index+1]):
            #remove current and next character
            text = text[:current_index]+text[current_index+2:]
            has_reaction = True
            break

print(text)      
```
